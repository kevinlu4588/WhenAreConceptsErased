```python
import torch
import torchvision.transforms as T
from torchvision import models
from PIL import Image
from transformers import CLIPProcessor, CLIPModel

# ============================================================
# üñºÔ∏è Load your image
# ============================================================
image_path = "/share/u/kevin/DiffusionConceptErasure/classifier_guidance/Screenshot 2025-10-23 at 10.15.04‚ÄØPM.png"  # <-- change this
image = Image.open(image_path).convert("RGB")

# ============================================================
# üß† ResNet-50 (ImageNet)
# ============================================================
resnet = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
resnet.eval()

# Transform for ResNet
preprocess_resnet = models.ResNet50_Weights.IMAGENET1K_V2.transforms()
img_tensor = preprocess_resnet(image).unsqueeze(0)

with torch.no_grad():
    logits = resnet(img_tensor)
    probs = torch.nn.functional.softmax(logits, dim=1)[0]

# Get top-5 predictions
labels = models.ResNet50_Weights.IMAGENET1K_V2.meta["categories"]
top5_prob, top5_catid = torch.topk(probs, 10)
print("\nüîç ResNet-50 Predictions:")
for i in range(10):
    print(f"{labels[top5_catid[i]]:<25} {top5_prob[i].item():.4f}")

# ============================================================
# ü§ù CLIP (zero-shot classifier)
# ============================================================
device = "cuda" if torch.cuda.is_available() else "cpu"
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Use ImageNet class names as text prompts for CLIP
text_inputs = [f"a photo of a {label}" for label in labels]

inputs = clip_processor(
    text=text_inputs,
    images=image,
    return_tensors="pt",
    padding=True
).to(device)

with torch.no_grad():
    outputs = clip_model(**inputs)
    logits_per_image = outputs.logits_per_image  # [1, 1000]
    probs = logits_per_image.softmax(dim=1).cpu()[0]

top5_prob, top5_catid = torch.topk(probs, 10)
print("\nüéØ CLIP Predictions:")
for i in range(10):
    print(f"{labels[top5_catid[i]]:<25} {top5_prob[i].item():.4f}")
```
